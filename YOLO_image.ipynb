```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os
```

```python
# Load the YOLOv11 model (small version)
model = YOLO("yolo11s.pt")  # Using the small model for balance between speed and accuracy
```

```python
# Download an image from the internet
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/60/Naxos_Taverna.jpg/1200px-Naxos_Taverna.jpg?20050319140129"  # Example image
image_path = "test_image.jpg"
!wget -O {image_path} {image_url}
```

```python
# Read the image
#OpenCV loads images in BGR format, so we convert them to RGB to match the expected color ordering of most visualization tools and deep learning models.
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Perform object detection
results = model(image)
```

```python
# Display the results
plt.imshow(results[0].plot())
plt.axis("off")
plt.show()
```
