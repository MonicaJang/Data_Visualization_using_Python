```python
seoul_cafe.loc[seoul_cafe['상호명'] == '이디야커피', '상호명'] = '이디야'
```

```python
seoul_cafe['상호명'].value_counts().head(10)    # 이디야로 통일 및 서울 상위 10개 커피 전문점 확인완료
```

```python
seoul_cafe['상호명'].value_counts().head(10).index
```

```python
top_ten = seoul_cafe[seoul_cafe['상호명'].isin(['이디야', '투썸플레이스', '메가엠지씨커피', '컴포즈커피', '매머드익스프레스', '더벤티', '빽다방', '커피에반하다', '카페인중독', '디저트39'])]
```

```python
top_ten
```

```python
import folium

# 중심 지정
lat = top_ten['위도'].mean()
long = top_ten['경도'].mean()
```

```python
m = folium.Map([lat, long], zoom_start=11)

# 지도위에 표시
for i in top_ten.index:
    sub_lat = top_ten.loc[i, '위도']
    sub_long = top_ten.loc[i, '경도']
    title = top_ten.loc[i, '상호명']
    
    #지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat, sub_long], tooltip = title).add_to(m)

m.save('top_ten_cafe.html')
m
```

```python
import json

geo_path = '02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
top_ten_count = pd.pivot_table(top_ten, index='시군구명', values='상가업소번호', aggfunc='count')
print(top_ten_count)
```

```python
m = folium.Map(location=[lat, long], zoom_start=11, tiles='OpenStreetMap', attr='© OpenStreetMap contributors')

folium.Choropleth(
    geo_data=geo_str,  # GeoJSON 데이터
    data=top_ten_count['상가업소번호'],  # 특정 지역의 데이터
    columns=['시군구명', '상가업소번호'],  # 사용할 데이터의 열
    fill_color='YlGn',  # 채우기 색상
    key_on='feature.properties.name',  # GeoJSON에서 지리적 개체를 식별하는 키
    legend_name='Number of cafe (%)'  # 범례 이름
).add_to(m)

m.save('top_ten_cafe.html')  # 지도를 HTML 파일로 저장
m
```
